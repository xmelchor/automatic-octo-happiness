Using **Bloom_etal_2018_Reduced_Dataset.csv**, we are going to do more dataframe manipulation and subsetting and summarizing

for example:

```python
#read in our .csv file
df=pd.read_csv('Bloom_etal_2018_Reduced_Dataset.csv')
df
```

```python
#splitting the taxa into individual words and putting them in columns 0 and 1
df['taxa'].str.split('_',n=1,expand=True)
```

```python
#adding the columns 'genus' and 'species' and splitting the taxa at underscore to put
#the individual word under them
df[['genus','species']]=df['taxa'].str.split('_',n=1,expand=True)
df
```

```python
#rearranging columns so 'genus' and 'species' are next to 'taxa'
cols = ['taxa','genus','species','logbodysize','trophic_position','Reg']
df[cols]
```

**2.2 Create a new dataframe with the mean *logbodysize* and *trophicposition* of each genera. Sort this data frame by the largest body size. Print the head of this dataframe. Which genera is the smallest and largest? What is the trophic position of the smallest and largest?**

```python
#Selcting only genus, logbodysize and trophicposition from the df and calculating the mean of latter two. 
mean_group_df = df.groupby('genus').agg('mean')
#renaming the logbodysize and trophic_position to mean_logbodysize and mean_trophicposition respectively
mean_logbobysize_df = mean_group_df.rename(columns={'logbodysize':'mean_logbodysize','trophic_position':'mean_trophicposition'})
mean_logbobysize_df
```

```python
#sort the dataframe by the largest body size
sorted_df = mean_logbobysize_df.sort_values(by=['mean_logbodysize'],ascending=False)
sorted_df.head()
```

```python
#largest genus according to logbodysize
max_logbodysize_genera = sorted_df[sorted_df['mean_logbodysize']==np.amax(sorted_df['mean_logbodysize'])]
max_logbodysize_genera
#the genus having maximum mean_logbodysize is Tenualosa with corresponding trophicposition of 0.462398
```

```python
#smallest genus according to logbodysize
min_logbodysize_genera = sorted_df[sorted_df['mean_logbodysize']==np.amin(sorted_df['mean_logbodysize'])]
min_logbodysize_genera
#the genus having minimum mean_logbodysize is Amazonsprattus with corresponding trophicposition of 0.531479
```

```python
import os.path
i=0
logfiles_dict = pd.DataFrame(columns = ('filename', 'minTemp', 'minDate','maxTemp', 'maxDate', 'meanTemp')) 
for f in logfiles: 
    #read one file
    infile = pd.read_csv(f,sep='\t',engine='python')
    #read filenames
    head, tail = os.path.split(f)
    os.path.basename(tail)
    filename = os.path.basename(tail).split('.',1)[0]
    #print(filename)
    infile.columns = ['Index','Date','Time','Temp','Type']
    infile = infile[['Date','Time','Temp']]
    #read min temperature
    minTemp = infile['Temp'].min()
    #print(minTemp)
    #read min temperature's date
    minDate = infile['Date'][infile['Temp'] == minTemp].unique()[0] 
    #print(minDate)
    #read max temperature
    maxTemp = infile['Temp'].max()
    #print(maxTemp)
    #read max temperature's date
    maxDate = infile['Date'][infile['Temp'] == maxTemp].unique()[0] 
    #print(maxDate)
    #calculate mean temperature
    meanTemp = infile['Temp'].mean()
    #print(meanTemp)
    #make a row information
    logfiles_dict.loc[i] = [filename, minTemp, minDate, maxTemp, maxDate, meanTemp]
    #print(logfiles_dict.loc[i])
    i+=1
#indexing the 'filename'
logfiles_dict = logfiles_dict.set_index('filename')
print(logfiles_dict)
```
